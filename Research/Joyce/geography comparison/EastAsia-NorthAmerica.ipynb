```python
# Install lifelines using: pip install lifelines
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Sample data (time to event, and event status)
time = [5, 10, 15, 20, 25]
event_status = [1, 1, 0, 1, 0]

# Create a Kaplan-Meier estimator
kmf = KaplanMeierFitter()

# Fit the estimator to the data
kmf.fit(time, event_observed=event_status)

# Plot the survival function
kmf.plot_survival_function()
plt.title("Kaplan-Meier Survival Function")
plt.xlabel("Time")
plt.ylabel("Survival Probability")
plt.show()
```

```python
# EastAsia follow-up month
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Load data from CSV file
file_path = '/Users/huyanshen/Desktop/EastAsia_follow_up.csv'
data = pd.read_csv(file_path)
data = data.dropna()

# Assuming your CSV has columns "time" and "event_status"
time_column = 'time'
event_status_column = 'event_status'

# Map 'Alive' to 1 and 'Disease' to 0
data[event_status_column] = data[event_status_column].map({'Alive': 0, 'Decease': 1})

# Create a Kaplan-Meier estimator
kmf = KaplanMeierFitter()

# Fit the estimator to the data
kmf.fit(durations=data[time_column], event_observed=data[event_status_column])

median_followup_time = data[time_column].median()
median_survival_time = kmf.median_survival_time_

ax = kmf.plot_survival_function()
ax.axvline(x=median_survival_time, color='red', linestyle='--',
           label=f'Median Survival Time: {median_survival_time:.2f} months')
ax.axvline(x=median_followup_time, color='green', linestyle='--',
           label=f'Median Follow-up Time: {median_followup_time:.2f} months')

# Plot the survival function
kmf.plot_survival_function()
plt.title("Kaplan-Meier Survival Function")
plt.xlabel("Time")
plt.ylabel("Survival Probability")
plt.legend()
plt.show()
```
