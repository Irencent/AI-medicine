```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_names = ['test_ecc_data.txt', 'train_ecc_data.txt', 'val_ecc_data.txt']
names = ['test_ecc_data', 'train_ecc_data', 'val_ecc_data']
```

```python
import os
result = []
for name,file_name in zip(names, file_names):
    with open(file_name, 'r') as file:
        for line in file:
            result.append(re.search('-\d+\.\d+', line).group())
    result_num = [float(i) for i in result] 

    plt.figure(figsize=(5, 4))
    sns.kdeplot(result_num, fill=True)
    plt.xlabel("label")
    plt.ylabel("text_ecc_data")
    figure_name = os.path.join('./Distribution',f"{file_name}.png")
    plt.savefig(figure_name)
    
    plt.figure(figsize=(5, 4))
    ax = sns.boxplot(x=result_num, y=None, linewidth=1, width=0.2, color='red')
    for spine in ["top", "right", "left"]:
        ax.spines[spine].set_visible(False)

    plt.xlabel(file_name)
    plt.ylabel("")

    figure_name = os.path.join('./Distribution',f"{name}.png")
    plt.savefig(figure_name)
      
```
