### 设定函数

```python
# 计算特征值之间是否具有关联
def corr_feature(df1,df2):
    feature_1 = df1.columns
    feature_2 = df2.columns
    
    temp_df = pd.merge(df1,df2,left_index=True,right_index=True,how='inner')
    corr_df = temp_df.corr()
    corr_df = corr_df.loc[feature_1,feature_2]
    
    return(corr_df)
```
